```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6,2),facecolor=(1, 1, 1) )
plt.xlim([0,11])
plt.ylim([-0.1,0.1])

# plt.cm.jet

for x in range(1,11):
    # circ = plt.Circle((x,0), 0.1, color=plt.cm.viridis(x/10))
    # circ = plt.Circle((x,0), 0.1, color=plt.cm.coolwarm(x/10))
    circ = plt.Circle((x,0), 0.1, color=plt.cm.bwr(x/10))
    ax.add_patch(circ)          
```
