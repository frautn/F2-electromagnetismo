## 1. <a id='toc1_'></a>[Campo de una carga](#toc0_)

El campo elctrico producido por una carga puntual $q$ ubicada en la posicin $\vec{\text{x}}_0$ puede escribirse como:
$$
\vec{E}(\vec{\text{x}}) = \frac{1}{4\pi\varepsilon_0} \frac{q}{|\vec{\text{r}}|^2} \hat{\textbf{r}}\, , \tag{1}
$$
donde
  * $|\vec{\text{r}}|$ es la distancia entre la posicin de la partcula cargada ($\vec{\text{x}}_0$) y la posicin donde se quiere obtener el vector de campo elctrico ($\vec{\text{x}}$)
  * $\hat{\text{r}}$ es el vector unitario en la direccin desde $\vec{\text{x}}_0$ hacia $\vec{\text{x}}$

segn se muestra en la siguiente figura.
<div style="text-align: center;">
       <image src="https://raw.githubusercontent.com/frautn/F2-electromagnetismo/main/modulos/img/em01_una_carga.svg">
</div>

Si el vector unitario se reescribe como $\hat{\textbf{r}} = \vec{\text{r}}/|\vec{\text{r}}|$, la expresin para el campo resulta:
$$
\vec{E}(\vec{\text{x}}) = \frac{1}{4\pi\varepsilon_0} \frac{q}{|\vec{\text{r}}|^3} \vec{\text{r}}\, . \tag{2}
$$

Para escribir un cdigo que calcule el campo elctrico es conveniente utilizar una expresin con las coordenadas del punto campo y las coordenadas de la posicin de la carga explcitas. El cdigo que utilizaremos reproduce la siguiente expresin:
$$
\vec{E}(x,y,z) = \frac{1}{4\pi\varepsilon_0} \frac{q}{\left ((x-x_0)^2 + (y-y_0)^2 + (z-z_0)^2 \right )^{3/2}} \left ( (x-x_0)\hat{\textbf{i}} + (y-y_0)\hat{\textbf{j}} + (z-z_0)\hat{\textbf{k}} \right ) . \tag{3}
$$

A continuacin de define una funcin que calcula las componentes del campo elctrico en la posicin $(x,y,z)$ cuando se tiene una carga de $1~\text{nC}$ en el origen:

```python
def E1(x,y,z):
    """Calcula las componentes del campo elctrico en N/C.
    Ingresar valores de x,y,z en metros y q en coulomb.
    """
    k = 9E9   #Constante de Coulomb en las unidades correspondientes.
    
    q = 1E-9
    x0 = 0
    y0 = 0
    z0 = 0

    # El denominador en la ec. 3:
    r32 = ((x - x0)**2 + (y - y0)**2 + (z - z0)**2)**(3/2)

    # Cada una de las componentes de la ec. 3:
    Ei = k * q * (x - x0) / r32
    Ej = k * q * (y - y0) / r32
    Ek = k * q * (z - z0) / r32

    return Ei, Ej, Ek
```

Si queremos el vector campo elctrico en alguna posicin, simplemente se ingresan los valores de $x,y,z$ (el resultado est expresado en N/C):

```python
E1(-1,0,0)
```

## 2. <a id='toc2_'></a>[Campo de un sistema de cargas puntuales](#toc0_)

El campo total producido por un sistema de $N$ cargas puntuales es la suma de los campos de cada carga (ver ecuacin (2) para el campo de una nica carga):
$$
\vec{E}(\vec{\text{x}}) = \frac{1}{4\pi\varepsilon_0} \sum_{i=1}^N \frac{q_i}{|\vec{\text{r}}_i|^3} \vec{\text{r}}_i\, , \tag{4}
$$

donde
* $q_i$ es la carga elctrica de la partcula i-sima 
* $\vec{\text{r}}_i$ es el vector diferencia entre la posicin donde se quiere obtener el vector de campo elctrico ($\vec{\text{x}}$) y la posicin de la partcula cargada ($\vec{\text{x}}_i$), es decir:
$$
\vec{\text{r}}_i = (x-x_i)\hat{\textbf{i}} + (y-y_i)\hat{\textbf{j}} + (z-z_i)\hat{\textbf{k}} ~ . \tag{5}
$$

### 2.1. <a id='toc2_1_'></a>[El dipolo elctrico](#toc0_)

Un dipolo es un sistema de dos cargas elctricas de igual valor absoluto pero signos opuestos, separadas por una distancia $d$. Para calcular el campo total podemos extender el cdigo utilizado anteriormente agregando los datos de la segunda carga y sumando, como en la definicin de la siguiente funcin:

```python
def E2(x,y,z):
    """Calcula las componentes del campo elctrico en N/C de un sistema de 2 cargas.
    Ingresar valores de x,y,z en metros y q en coulomb.
    """
    k = 9E9   #Constante de Coulomb en las unidades correspondientes.

    # Datos de la carga 1
    q1 = -1E-9
    x1 = -2
    y1 = 0
    z1 = 0

    # Datos de la carga 2
    q2 = 1E-9
    x2 = 2
    y2 = 0
    z2 = 0

    r = ((x - x1)**2 + (y - y1)**2 + (z - z1)**2)**(3/2)
    Ei1 = k * q1 * (x - x1) / r
    Ej1 = k * q1 * (y - y1) / r
    Ek1 = k * q1 * (z - z1) / r

    r = ((x - x2)**2 + (y - y2)**2 + (z - z2)**2)**(3/2)
    Ei2 = k * q2 * (x - x2) / r
    Ej2 = k * q2 * (y - y2) / r
    Ek2 = k * q2 * (z - z2) / r

    return Ei1+Ei2, Ej1+Ej2, Ek1+Ek2
```

Verifiquemos que nuestro cdigo funciona correctamente, calculando los vectores E en diferentes posiciones. En el origen, el campo apunta hacia la izquierda:

```python
E2(0,0,0)
```

A la misma distancia a ambos lados del dipolo obtenemos el mismo vector:

```python
E2(4,0,0) , E2(-4,0,0)
```

A la misma distancia en $y$ y en $z$ en el centro obtenemos el mismo vector:

```python
E2(0,2,0) , E2(0,0,2)
```

Otras posiciones:

```python
E2(2,2,0) , E2(-2,2,0) , E2(0,2,2)
```
